```python
from tokenizers import ByteLevelBPETokenizer
from datasets import load_dataset
import os

# 1. 'Glot500' dataset use kar rahe hain - ye multilingual ke liye best hai
# Isme Hindi (hin_Deva) ka data direct uthayenge
print("Data load ho raha hai... is baar pakka chalega! ðŸš€")

try:
    # Hindi data load kar rahe hain
    dataset = load_dataset("cis-lmu/Glot500", "hin_Deva", split="train[:5000]", trust_remote_code=True)

    # Ek text file banao training ke liye
    with open("train_data.txt", "w", encoding="utf-8") as f:
        for item in dataset:
            f.write(item['text'] + "\n")

    # 2. Tokenizer ko Train karo
    tokenizer = ByteLevelBPETokenizer()
    tokenizer.train(files=["train_data.txt"], vocab_size=30000, min_frequency=2,
                    special_tokens=["<s>", "<pad>", "</s>", "<unk>", "<maskç´§"])

    # 3. Save karo
    os.makedirs("desi-tokenizer", exist_ok=True)
    tokenizer.save_model("desi-tokenizer")
    print("Mubarak ho! Tokenizer 'desi-tokenizer' folder mein save ho gaya hai. ðŸ˜Ž")

except Exception as e:
    print(f"Error aa gaya bhai: {e}")
```

```python
from transformers import GPT2Config, GPT2LMHeadModel, GPT2TokenizerFast

# 1. Model ki configuration set karo (Chhota aur Tej)
config = GPT2Config(
    vocab_size=30000,       # Jo tokenizer mein rakha tha
    n_positions=512,        # Ek baar mein kitne words yaad rakhega
    n_embd=256,             # Embedding ki depth (Dimensions)
    n_layer=6,              # Kitni layers hongi (Brain depth)
    n_head=8,               # Attention heads (Kitni jagah dhyan dega)
)

# 2. Model ko initialize karo (Scratch se)
model = GPT2LMHeadModel(config)

# 3. Pehle se bana Tokenizer load karo
tokenizer = GPT2TokenizerFast.from_pretrained("./desi-tokenizer",
                                            bos_token="<s>",
                                            eos_token="</s>",
                                            pad_token="<pad>",
                                            unk_token="<unk>")

print(f"Bhai, Model tyar hai! Isme total {model.num_parameters():,} parameters hain.")
```

```python
from transformers import DataCollatorForLanguageModeling, Trainer, TrainingArguments
from datasets import load_dataset

# 1. Dataset ko Tokenize karo
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, max_length=512)

raw_dataset = load_dataset("text", data_files={"train": "train_data.txt"})
tokenized_dataset = raw_dataset.map(tokenize_function, batched=True, remove_columns=["text"])

# 2. Data Collator
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

# 3. Training Arguments (Fix kiya hua)
training_args = TrainingArguments(
    output_dir="./desi-ai-model",
    num_train_epochs=3,           # Data ko 3 baar padhega
    per_device_train_batch_size=8,
    save_steps=500,
    save_total_limit=2,
    prediction_loss_only=True,
    logging_steps=100,
    report_to="none"              # Extra logs band karne ke liye
)

# 4. Trainer Initialize karo
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=tokenized_dataset["train"],
)

# 5. Shuru karo Training!
print("Training shuru ho rahi hai... Is baar pakka chalega! ðŸš€")
trainer.train()
```

```python
from transformers import DataCollatorForLanguageModeling, Trainer, TrainingArguments, GPT2Config, GPT2LMHeadModel, GPT2TokenizerFast
from datasets import load_dataset
from tokenizers import ByteLevelBPETokenizer
import os

# 1. Bada Data Load Karo (50k lines)
print("Bada dataset load ho raha hai... thoda time lagega â³")
dataset = load_dataset("cis-lmu/Glot500", "hin_Deva", split="train[:50000]", trust_remote_code=True)

with open("train_data_large.txt", "w", encoding="utf-8") as f:
    for item in dataset:
        f.write(item['text'] + "\n")

# 2. Tokenizer ko re-train karo bade data pe
tokenizer_raw = ByteLevelBPETokenizer()
tokenizer_raw.train(files=["train_data_large.txt"], vocab_size=30000, min_frequency=2,
                special_tokens=["<s>", "<pad>", "</s>", "<unk>", "<mask>"])
os.makedirs("desi-tokenizer-v2", exist_ok=True)
tokenizer_raw.save_model("desi-tokenizer-v2")

tokenizer = GPT2TokenizerFast.from_pretrained("./desi-tokenizer-v2",
                                            bos_token="<s>", eos_token="</s>",
                                            pad_token="<pad>", unk_token="<unk>")

# 3. Model Configuration (Wahi purani 12.5M wali, but fresh)
config = GPT2Config(vocab_size=30000, n_positions=512, n_embd=256, n_layer=6, n_head=8)
model = GPT2LMHeadModel(config)

# 4. Tokenize and Prepare Data
def tokenize_func(examples):
    return tokenizer(examples["text"], truncation=True, max_length=512)

raw_ds = load_dataset("text", data_files={"train": "train_data_large.txt"})
tokenized_ds = raw_ds.map(tokenize_func, batched=True, remove_columns=["text"])
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

# 5. Heavy Training Args
training_args = TrainingArguments(
    output_dir="./desi-ai-v2",
    num_train_epochs=10,           # 10 baar poora data padhega
    per_device_train_batch_size=16,
    fp16=True,                      # GPU Speed Booster! âš¡
    save_steps=1000,
    logging_steps=100,
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=tokenized_ds["train"],
)

print("Ab asli training shuru! ðŸ”¥")
trainer.train()
```
